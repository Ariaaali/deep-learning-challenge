## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from sklearn.ensemble import RandomForestClassifier

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
application_df_clean = application_df.drop(['EIN', 'NAME'], axis=1)
application_df_clean
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
application_df_clean.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
#  YOUR CODE GOES HERE
unique_count= application_df_clean['APPLICATION_TYPE'].value_counts()
unique_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE
type_counts = application_df_clean['APPLICATION_TYPE'].value_counts()
application_types_to_replace = type_counts[type_counts < 500].index
# Replace in dataframe
for app in application_types_to_replace:
    application_df_clean['APPLICATION_TYPE'] = application_df_clean['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df_clean['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
#  YOUR CODE GOES HERE
unique_count_classification= application_df_clean['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
unique_count_classification_above_one = unique_count_classification[unique_count_classification >1]
unique_count_classification_above_one
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
classifications_to_replace = unique_count_classification[unique_count_classification <1000].index
# Replace in dataframe
for cls in classifications_to_replace:
        application_df_clean['CLASSIFICATION'] = application_df_clean['CLASSIFICATION'].replace(cls,"Other")


# Check to make sure replacement was successful
application_df_clean['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
application_df_encoded = pd.get_dummies(application_df_clean)
application_df_encoded_df = application_df_encoded.astype(int)
application_df_encoded_df
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
y = application_df_encoded_df["IS_SUCCESSFUL"]
X = application_df_encoded_df.drop(["IS_SUCCESSFUL"], axis=1)
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(X, y)
```

```python
importances = rf_model.feature_importances_
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```

```python

application_df_encoded_new = X.drop(
    ['AFFILIATION_Other', 'ORGANIZATION_Corporation', 'AFFILIATION_National','AFFILIATION_Family/Parent', 'INCOME_AMT_5M-10M','USE_CASE_Heathcare', 'APPLICATION_TYPE_Other','INCOME_AMT_50M+', 'SPECIAL_CONSIDERATIONS_N', 'SPECIAL_CONSIDERATIONS_Y','STATUS','USE_CASE_Other'],
    axis=1

)
```

```python
application_df_encoded_new
```

```python
y1 = y
X1=application_df_encoded_new.values
```

```python
# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, random_state=78)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X1_scaler = scaler.fit(X1_train)

# Scale the data
X1_train_scaled = X1_scaler.transform(X1_train)
X1_test_scaled = X1_scaler.transform(X1_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()
number_input_features = len(X1_train[0])
hidden_nodes_layer1 =  20
hidden_nodes_layer2 = 10

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))
# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
#  YOUR CODE GOES HERE
checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath='model_weights_{epoch:02d}.weights.h5',
    save_weights_only=True,
    save_freq=5
)
fit_model=nn.fit(X1_train_scaled, y1_train, epochs=100, callbacks=[checkpoint_callback])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X1_test_scaled,y1_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
